```sas
PROC MEANS data=SASHELP.HEART;
RUN;
```

```sas
TITLE "Statistics on Cholesterol based on weight status";
    PROC MEANS data=SASHELP.HEART N MEAN STDDEV STDERR CLM;
    CLASS WEIGHT_STATUS;
RUN;
```

```sas
ODS GRAPHICS ON;
TITLE " Histogram"
PROC UNIVARIATE data=SASHELP.HEART;
    CLASS WEIGHT_STATUS;
    HISTOGRAM CHOLESTEROL;
    INSET SKEWNESS KURTOSIS;
    PROBPLOT CHOLESTEROL;
    INSET SKEWNESS KURTOSIS; 
RUN;
    
ODS GRAPHICS OFF;
```
